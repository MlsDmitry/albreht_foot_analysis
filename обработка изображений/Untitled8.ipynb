```python
def split_foot(image):
    leftPad = 280
    rightPad = 1920 - 1630
    assert image.shape == (1080, 1920, 3)
    return image[:,leftPad:880,:], image[:,1025:-rightPad,:]
    
```

```python
footL, footR = split_foot(image)
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(footL)
ax2.imshow(footR)
print(footL.shape, footR.shape)
```
